```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

```python
#dropping species column
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
#adding sepal_sum column
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
#adding petal_sum column
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
mean = [np.mean(iris['sepal_length']),
        np.mean(iris['sepal_width']),
        np.mean(iris['sepal_sum']),
        np.mean(iris['petal_length']),
        np.mean(iris['petal_width']),
        np.mean(iris['petal_sum'])]

summary_dict = {'mean':mean}

summary_df = pd.DataFrame(data = summary_dict)
summary_df
```

```python
#adding labels, std, min and max to mean
my_dict = {'mean' :  [np.mean(iris['sepal_length']),
                     np.mean(iris['sepal_width']),
                     np.mean(iris['sepal_sum']),
                     np.mean(iris['petal_length']),
                     np.mean(iris['petal_width']),
                     np.mean(iris['petal_sum'])],
           'std'  : (iris.std()),
           'min' : (iris.min()),
           'max' : (iris.max())}

summary_df = pd.DataFrame (my_dict, index = ['sepal_length', 'sepal_width','sepal_sum','petal_length', 'petal_width', 'petal_sum'])

summary_df
```
